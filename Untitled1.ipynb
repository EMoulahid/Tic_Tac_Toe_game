# Tic-Tac-Toe game in Python



```python
from enum import Enum

class PlayerIcon(Enum):
    X = "X"
    O = "O"

def print_board():
    print("\nBoard:")
    for i in range(0, 9, 3):
        print("| {} | {} | {} |".format(board[i], board[i + 1], board[i + 2]))
    print()

def player_move(player_icon):
    print("Your turn, player {}".format(player_icon.value))
    while True:
        try:
            choice = int(input("Enter your move (1-9): ").strip())
            if 1 <= choice <= 9 and board[choice - 1] == " ":
                board[choice - 1] = player_icon.value
                break
            else:
                print("Invalid move. Please choose an empty space (1-9).")
        except ValueError:
            print("Invalid input. Please enter a number.")

def check_game_status(player_icon):
    win_conditions = [
        (0, 1, 2),  # Top row
        (3, 4, 5),  # Middle row
        (6, 7, 8),  # Bottom row
        (0, 3, 6),  # Left column
        (1, 4, 7),  # Middle column
        (2, 5, 8),  # Right column
        (0, 4, 8),  # Diagonal \
        (2, 4, 6)   # Diagonal /
    ]
    
    for condition in win_conditions:
        if all(board[i] == player_icon.value for i in condition):
            print_board()
            print(f"{player_icon.value} wins! Congratulations!")
            return True
    
    if " " not in board:
        print_board()
        print("It's a draw!")
        return True

    return False

# Initialize the game board
board = [" " for _ in range(9)]

# Main game loop
while True:
    print_board()
    player_move(PlayerIcon.X)
    if check_game_status(PlayerIcon.X):
        break
    print_board()
    player_move(PlayerIcon.O)
    if check_game_status(PlayerIcon.O):
        break
```
